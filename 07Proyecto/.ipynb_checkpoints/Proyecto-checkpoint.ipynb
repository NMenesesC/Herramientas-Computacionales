#### Primera Parte detección de la placa

Primero se implementa la función *cargar* que permite leer los archivos de imagenes almacenados en la dirección/carpeta del proyecto. En dicha función se hace un pre-procesamiento de la imagen con el fin de facilitar la detección de contornos, bordes o particularmente rectángulos. Como primera fase se tiene el recorte de la imagen tomando la porción en la cual tiende a encontrarse la matrícula de un vehículo. Se elimina el 25% de la imagen en su parte superior en un sentido vertical, además se reduce a un 50% horizontalmente centrando la porción a visualizar. 

Por otra parte se tiene el procesamiento o filtrado de la imagen mediante la función de Canny, la cual a partir de un contraste de colores identifica los contornos más drásticos de la imagen y los binariza con el fin de obtener un panorama claro de las figuras presentes en la imagen. Al final, la función plotea la imagen con el filtro de Canny. 

```python
#Lectura imagen de prueba
def cargar(nombre):
    img = cv2.imread('Imagenes/'+ nombre)
    ly, lx = img.shape[0:2]
    lx25=lx*0.2
    ly25=ly*0.25
    lend=lx-lx25
    # Cropping
    cort= img[int((ly/2)-ly25):,int((lx/2)-lx25):int((lx/2)+lx25), :]
    edged = cv2.Canny(cort, 200, 300)
    figure()
    pylab.imshow(edged, 'gray')
    return edged, cort
```

Ahora bien, se contruye la función *deteccion* que a partir de la imagen binarizada obtenida en la función *cargar* detecta todos los posibles rectangulos descritos por los bordes. Estos "contornos" detectados son organizados y filtrados con el fin de ubicar el que mejor se adecúa a la matrícula del vehículo. 

Como primer paso se organizan de mayor a menor área de contornos, y se reduce la cantidad de resultados en un 80% despreciando los de menor área). A partir de los contornos restantes se crea un bucle que evalúa mediante condicionales las carácteristicas de cada contorno, con el fin de seguir filtrando los resultados. Como condición se tiene la relación entre el ancho y alto del rectángulo,  ya que gracias a lo investigado se sabe que una matrícula de vehículo colombiano tiene una relación entre 1.8 y 2.2. 

Finalmente los contornos filtrados se dibujan en la imagen, para que el usuario valide el proceso de selección de los rectángulos. 

```python
def deteccion(edged, cort):  
    img2, contornos, hierarchy = cv2.findContours(edged, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    k = len(contornos)*0.8
    contornos = sorted(contornos, key = cv2.contourArea, reverse = True)[:int(k)]
    screenCnt = None
    cont = list()
    for i in range(0,len(contornos)):
        x,y,w,h = cv2.boundingRect(contornos[i])
        if h>0 and w/h > 4.5 and w/h < 7 and h < 80 and w < 300:
            cort= cv2.rectangle(cort,(x,y),(x+w,y+h),(0,150,0),2) 
            rect = cv2.minAreaRect(contornos[i])
            cont.append(contornos[i])    
    figure()
    pylab.imshow(cort)
    return cont
```

Como última fase de filtrado se tiene la función *placa*. En esta función se busca el contorno más grande, el cual tiende a ser en el que se ubica la matrícula. A partir de el contorno seleccionado se procede a colorear todos los pixeles al rededor de la placa (con una tolerancia de 3 pixeles) de color negro. Esto con el fin de en las siguientes fases de detección de caracteres tener únicamente la información necesaria, eleminando la mayor cantidad de ruido en la imagen. Esta función arroja como resultado la imagen completamente filtrada.    

```python
def placa(cont, nombre):
    l = 0
    a = 0
    for j in range(0,len(cont)):
        x,y,w,h = cv2.boundingRect(cont[j])
        if w > l:
            l = w
            a = j    

    x,y,w,h = cv2.boundingRect(cont[a])
    img = cv2.imread('Imagenes/'+nombre)
    ly, lx = img.shape[0:2]
    lx25=lx*0.2
    ly25=ly*0.25
    lend=lx-lx25
    cort= img[int((ly/2)-ly25):,int((lx/2)-lx25):int((lx/2)+lx25), :]
    cort= cort[y:y+h,x:x+w,:]
    #cort[:y-3,:,:] = [0,0,0]
    #cort[:,:x-3,:] = [0,0,0]
    #cort[y+h+3:,:,:] = [0,0,0]
    #cort[:,x+w+3:,:] = [0,0,0]
    
    gray = cv2.cvtColor(cort,cv2.COLOR_BGR2GRAY)
    figure()
    pylab.imshow(gray, 'gray')
    return gray
```

A continuacion se expone un ejemplo con una de las imagenes almacenadas en la carpeta:

```python
nombre = 'bmwUE.jpg'
edged, cort = cargar(nombre)
cont = deteccion(edged,cort)
gray = placa(cont, nombre)
ret,thresh = cv2.threshold(gray,0,250,cv2.THRESH_OTSU)
kernel = np.ones((1,1),np.uint8)
opening = cv2.morphologyEx(thresh,cv2.MORPH_OPEN,kernel, iterations = 2)
figure()
pylab.imshow(opening, 'gray')
```
