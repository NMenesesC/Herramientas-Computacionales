## Ejercicio 5.5

```python
%pylab inline
from scipy.optimize import curve_fit

x = np.array([3,4,5,7,8,9,11,12])
y = np.array([1.6,3.6,4.4,3.4,2.2,2.8,3.8,4.6])

def fun(x,a,b,c,d):
    return a*x**3+b*x**2+c*x+d

popt, pcov = curve_fit(fun, x, y)

print(popt)
```

```python
xp = np.linspace(x[0], x[-1]) 

plt.plot(x,y, '*r', label='datos')
plt.plot(xp,fun(xp, *popt), '-b', 
         label='fit: $a$=%.2f, $b$=%.2f, $c$=%.2f, $d$=%.2f' 
         % (popt[0], popt[1], popt[2], popt[3]))
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
```
