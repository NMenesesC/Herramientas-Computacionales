```python
%pylab inline
import matplotlib.pyplot as plt
import numpy as np
```

#### Ejemplo 1

```python
L = 20
n = 128
x2 = np.linspace(-L/2,L/2,n+1)
x = np.array(x2[0:n])
u = np.exp((-x)*x)
ut = np.fft.fft(u)
utshift = np.fft.fftshift(ut)

figure(figsize=(15,10))
plt.subplot(2,2,1)
plt.plot(x,u)
plt.subplot(2,2,2)
plt.plot((ut))
plt.subplot(2,2,3)
plt.plot((np.abs(utshift)))
```
